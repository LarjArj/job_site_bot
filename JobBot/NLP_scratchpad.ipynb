```python
word_frequencies ={}
for word in doc:
    
    unwanted = {"\n","\n\n",":","/",",",".",")","(","-",'”',"details",
                "Internship","Full","job","Job"}
    if word.text.lower() not in stopwords and word not in unwanted:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text]=1
            else:
                word_frequencies[word.text]+=1
```

```python
skills = {}
for word in word_frequencies:
    
    if word[0].isupper() and word not in skills:
        skills[word] = 1
    elif word[0].isupper() and word in skills:
        skills[word] +=1

skills
```
